# Evaluate focal_resolver performance

Script to evaluate the results of the focal_resolver_tool by comparing it to manual assignment of focal-nonfocal.
The evaluation was made for a random selection of N=150 calls, that the tool has assigned focal and N=150 that the tool has assigned nonfocal. 
Calls have been presented to Vlad, Baptiste and me and we labelled by hand whether a) this call has a match among its potential match partners and b) whether we would assign the call of interest as focal or nonfocal. These choices are then compared to the predictions of the tool (saved in pred_labelfile)

```python
# Functions to calculate, print and plot evaluation metrics of predictions

def calc_confusion_vals(pred, truth):
    
    eval_dict = {}
    eval_dict['TP'] = np.sum([1 if t==1 and p==1 else 0 for t,p in zip(truth, pred)])
    eval_dict['TN']= np.sum([1 if t==0 and p==0 else 0 for t,p in zip(truth, pred)])
    eval_dict['FP'] = np.sum([1 if t==0 and p==1 else 0 for t,p in zip(truth, pred)])
    eval_dict['FN'] = np.sum([1 if t==1 and p==0 else 0 for t,p in zip(truth, pred)])    
    return eval_dict


def get_metric(metric, eval_dict):
    
    TP = eval_dict['TP']
    TN = eval_dict['TN']
    FP = eval_dict['FP']
    FN = eval_dict['FN']
       
    
    if metric in ["Sensitivity", "sensitivity", "recall", "Recall", "TPR", "tpr"]:
        res = TP/(TP+FN)
    elif metric in ["Accuracy", "accuracy", "acc"]:
        res = (TP+TN)/(TP+TN+FN+FP) 
    elif metric in ["Specificity", "specificity", "TNR", "tnr"]:                   
        res = TN/(TN+FP)
    elif metric in ["Precision", "precision", "PPV", "ppv"]:
        res = TP/(TP+FP)
    elif metric in ["NPV", "npv"]:
        res = TN/(TN+FN) 
    elif metric in ["F1", "f1", "f1score", "F1Score"]:
        res = (2*TP)/(2*TP+FP+FN)
    elif metric in ["FPR", "fpr"]:
        res = 1-(TN/(TN+FP))
    else:
        print("unknown metric")
    
    return res

def print_metrics(eval_dict):
    
    # sensitivity/true positive rate/recall
    # -->What percentage of 1 were predicted as 1?
    sensitivity = get_metric("sensitivity", eval_dict)
    print("Sensitivity/TPR: ", round(sensitivity*100,2))

    # precision/PPV
    # --> What percentage of predicted 1 are actually 1?
    precision = get_metric("precision", eval_dict)
    print("Precision/PPV: ", round(precision*100,2))

    # specificity/
    # -->What percentage of predicted 0 are actually 0?
    specificity = get_metric("specificity", eval_dict)
    print("Specificity/TNR: ", round(sensitivity*100,2))


    # accuracy - "true predictions among all"
    # --> Of all cases, how many were predicted correctly?
    accuracy = get_metric("accuracy", eval_dict)
    print("Accuracy: ", round(accuracy*100,2))
    
    return None
```

```python
# Read in all answers 
labellers = ['vlad', 'baptiste', 'mara']

answer_df_dict={}

for labeller in labellers:
    RESULTS_FOLDER = ANSWERS+labeller+"/"
    answer_txts = sorted(glob(RESULTS_FOLDER+'*.txt'))
    
    answer_list = []

    for txt in answer_txts:
        answer_list.append(pd.read_csv(txt, sep=";", header=None))
    answer_df = pd.concat(answer_list)
    answer_df.columns = ['callID_new', 'match', 'nonfocal']
    
    
    # correct double clicks:
    answer_df=answer_df.replace('focalfocal', 'focal')
    answer_df=answer_df.replace('nonfocalnonfocal', 'focal')
    answer_df.nonfocal.value_counts()
    
    if answer_df.match.value_counts().shape[0]==2 and answer_df.nonfocal.value_counts().shape[0]==2:
        # remove duplicates
        answer_df = answer_df.drop_duplicates()      
        
        answer_df['nonfocal'] = [1 if x=="nonfocal" else 0 for x in answer_df.nonfocal]
        answer_df['match'] = [1 if x=="match" else 0 for x in answer_df.match]
        
        answer_df_dict[labeller] = answer_df
    else:
        print("Too many levels in "+labeller)
```

## Metric Comparison plot

```python
def plot_metric(stat_df, metric, plot_type="full"):
    
                   
    if plot_type=="full":        
        mask = np.full(statstab.shape, False)
        
    else:
        mask = np.full(statstab.shape, False)
        for i in range(mask.shape[0]):
            for j in range(i,mask.shape[0]):
                mask[j,i] = True
    
    fig, ax = plt.subplots(figsize=(5,5))         
    p1 = sns.heatmap(stat_df,
                     cmap="YlGn",
                     mask = mask,
                     annot=True,
                     cbar=False,
                     square=True,
                     linewidths=0.01)
    plt.title(metric)
    
    return fig
```

```python
#metric = 'Specificity'
param = 'nonfocal'
labellers = ['algo', 'baptiste','vlad', 'mara']


for metric in ['Accuracy','Specificity', 'Sensitivity', 'NPV', 'Precision']:
    statstab = np.zeros((len(labellers), len(labellers)))

    for i,key in enumerate(labellers):
        pred = list(answer_df_dict[key][param])
        for j,other_key in enumerate(labellers):
            truth = list(answer_df_dict[other_key][param])        
            eval_dict = calc_confusion_vals(pred, truth)
            statstab[i,j] = get_metric(metric, eval_dict) 

    # make dataframe
    stat_df = pd.DataFrame(statstab, index=labellers, columns=labellers)

    stat_df.index.name = "Prediction"
    stat_df.columns.name = "Truth"

    if metric in ['accuracy', 'Accuracy', 'acc']:
        p= plot_metric(stat_df, metric, "half")
    else:
        p=plot_metric(stat_df, metric)

    p.savefig("figures/"+metric+"_"+param+".jpg")
```

```python
humans = ['vlad', 'baptiste', 'mara']
algorithm = 'algo'

for param in ['_nf', '_match']:
    print(param,":")
    #param = "_match"

    # Find calls (rows) where humans agree
    # can be either of these cases: 
    # - sum of human predictions is len(humans) (all say 1)
    # - sum of human predictions is 0           (all say 0)

    l=[]
    for i in range(nonfocal_df.shape[0]):
        vals = []
        for human in humans:
            vals.append(nonfocal_df.loc[i,human+param])
        l.append(vals)

    human_agreement = [1 if (np.sum(x)==len(humans) or np.sum(x)==0) else 0 for x in l]
    nonfocal_df['human_agreement'+param] = human_agreement

    # all calls where humans agree
    human_agree_df = nonfocal_df.loc[nonfocal_df['human_agreement'+param]==1,:]
    print(human_agree_df.shape)
    
    # What was the human choice?
    # can choose any human, as they all agree. By default will choose first in list
    human_choice = human_agree_df[humans[0]+param]
    human_agree_df['human_choice'+param] = human_choice    
    
    # Add choice to nonfocal_df
    human_choice_dict = dict(zip(human_agree_df['callID_new'], human_agree_df['human_choice'+param]))
    choice = [human_choice_dict[x] if x in human_choice_dict.keys() else "NA" for x in nonfocal_df.callID_new]
    nonfocal_df['human_choice'+param] = choice   
```

### What is delta intensity for correctly assigned focal-nonfocal pairs vs. incorrectly assigned focal-nonfocal pairs?

```python
# What is delta intensity for correctly assigned focal-nonfocal pairs vs. incorrectly assigned focal-nonfocal pairs

# select all calls ALL humans AND algorithm agree there is a match
agreed_matches_df = nonfocal_df.loc[((nonfocal_df['human_choice_match']==1) & (nonfocal_df[algorithm+'_match']==1)),:]
print(agreed_matches_df.shape[0])

# select all calls that have only one match (reduces data, but easier to analyze)
single_match = []
dist_scores = []
deltas = []

for call in agreed_matches_df.callID_new.values:    
    sub_df = f_nf.loc[(f_nf['call_a']==call) | (f_nf['call_b']==call),:]        
    sub_df_matched = sub_df.loc[sub_df.dist_score<CUTOFF,:]   
    
    # if call has exactly 1 match:
    if sub_df_matched.shape[0]==1:
        single_match.append(call)
        dist_scores.append(sub_df_matched.dist_score.values[0])
        deltas.append(np.abs(sub_df_matched.intense_a.values[0]-sub_df_matched.intense_b.values[0]))
print(len(single_match))
            
agreed_and_single_matches_df = agreed_matches_df.loc[agreed_matches_df['callID_new'].isin(single_match),:] 
agreed_and_single_matches_df['dist_score'] = dist_scores
agreed_and_single_matches_df['delta'] = deltas

# delta intensity of match
# select all calls where there is full agreement between nonfocal choices of all humans and the algorithm
# plot their delta intensity in histogram as "clear and correct cases"
full_agreement = [1 if x==y else 0 for x,y in zip(agreed_and_single_matches_df['human_choice_nf'], 
                                                  agreed_and_single_matches_df[algorithm+'_nf'])]
agreed_and_single_matches_df['full_agreement'] = full_agreement

# select all calls where there is full agreement between nonfocal choices of all humans, but algorithm disagrees
# plot their delta intensity in histogram as "clear but incorrect cases"
clear_different = [1 if (x==1 and x!=y) else 0 for x,y in zip(agreed_and_single_matches_df['human_agreement_nf'], 
                                                              agreed_and_single_matches_df[algorithm+'_nf'])]
agreed_and_single_matches_df['clear_different'] = clear_different

# select all other calls 
# plot their delta intensity in histogram as "unclear cases (labeller disagreement)"
none_of_either = [1 if (x==0 and y==0) else 0 for x,y in zip(agreed_and_single_matches_df['full_agreement'], 
                                                            agreed_and_single_matches_df['clear_different'])]
agreed_and_single_matches_df['none_of_either'] = none_of_either
```
